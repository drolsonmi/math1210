The first thing we need to do is set up our variable(s) and function(s):

````python
x = var('x')
f(x) = x^2 + 2*x + 1
````

### Basic Plots
The simplest way to graph a function is just to use the following command:

````python
plot(f(x))
````

However, there is another method that will be more useful. We do this by saving the plot to a variable. Doing this, we can make simple edits to the plot.

````python
p = plot(f(x))
p.show()
````

### Labeling the Graph
NEVER should you submit a graph without the following
* A title
* Axis labels
* A legend

To do this, we will take advantage of the saved plot. We can add a title by adding `.title("Your Title")` to the saved plot. Likewise, we add `axes_labels=["x variable","y variable"]` to the plot command.

```python
p = plot(f(x), title="Position as a function of Time", axes_labels=["Time (s)","Position (m)"])
```

As we add more and more information to the plot, this command can get long and cumbersome. I recommend just adding new commands to a new line before closing parenthesis (that is, wherever you have a comma, just press `Enter` for the new command).

````python
p = plot(f(x),
         title="Position as a function of Time",
         axes_labels=["Time (s)","Position (m)"])
p.show()
````

(Sometimes, the y label and the title overlap each other, depending on the computer. There are other ways to solve this, but we don't need to get into that right now - maybe another day. For now, just pick one or the other.)

### Adjust the Ranges
Now, let's adjust the view of the plot. Let's say that we want to view this plot for $-0.5\le x \le 0.5$ and $0\le y \le 2.0$. To do this, we just add `(x,-0.5,0.5)`. Alternatively, you can add `xmin=-0.5, xmax=0.5` to the plot function.
* Note that for the y range, you can only use `ymin=` and `ymax=`.

````python
p = plot(f(x),
         (x,-0.5,0.5),
         ymin=0, ymax=2.0,
         title="Position (m) as a function of Time (s)",
         axes_labels=["Time (s)",""])
p.show()
````

### Styling the Plot
Often, figures will need to display more than one function. But when you do, how do you distinguish one graph from the other? There are two ways to do this:
1. Graph with different colors
2. Graph with different patterns

To change the color of a line, add `color=red` to the command. Replace 'red' with whatever color you want to use.

To change the pattern of line, add one of the following:
1. `linestyle="-"` for a solid line
2. `linestyle="--"` for a dashed line
3. `linestyle=":"` for a dotted line
4. `linestyle="-."` for alternating dash and dot

````python
g(x) = sin(10*x)

pg = plot(g(x),
          color='red',
          linestyle="-.")
pg.show()
````

### Multiple Lines on One Plot
Often, we need to show two lines in the same plot. In this exercise, we have already defined two functions:
$$f(x) = x^2 + 2x + 1 \qquad g(x)=(\sin(7x))^2$$

If we want to show the two plots together, we can simply use an addition operator:
```python
pf = plot(f(x))
pg = plot(g(x))
p = pf + pg
p.show()
```

However, when you do have multiple lines in a single plot, it is important to label each line. We now add the  `legend_label="f(x)"` line to our command. Plotting both figures, we get:

````python
pf = plot(f(x),
         (x,-0.5,0.5),
         title="Position (m) as a function of Time (s)",
         axes_labels=["Time (s)",""],
         legend_label="f(x) (an accelerating car)")

pg = plot(g(x),
          xmin=-0.5, xmax=0.5,
          color='red',
          linestyle="-.",
          legend_label="g(x) (a spring)")

p = pf + pg
p.show()
````
